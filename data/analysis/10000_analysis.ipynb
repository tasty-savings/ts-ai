```python
import pandas as pd
import ast

# CSV 파일 읽기
df = pd.read_csv('../result/10000recipe.csv')

# ingredients 컬럼을 문자열에서 리스트로 변환
df['ingredients'] = df['ingredients'].apply(ast.literal_eval)

# cooking_order 컬럼을 문자열에서 리스트로 변환
df['cooking_order'] = df['cooking_order'].apply(ast.literal_eval)

print(df.dtypes)
df
```

### 레시피 인분에 따른 통계

```python
# 기초 통계 정보 확인
servings_stats = df['servings'].describe()
print(servings_stats)

# 고유 값 확인
unique_servings = df['servings'].unique()
print(f"Unique servings values: {sorted(unique_servings)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(df['servings'], bins=30, kde=True)
plt.title('Distribution of Servings')
plt.xlabel('Servings')
plt.ylabel('Frequency')

# 파일 저장
plt.savefig('../img/10000_Distribution of Servings.png')  

plt.show()
```

## 레시피 인분별 조회수

```python
import matplotlib.pyplot as plt

# servings별 views의 평균 계산
grouped_data = df.groupby('servings', as_index=False)['views'].mean()
grouped_data
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['servings'].astype(str), grouped_data['views'], color='skyblue')
plt.title('Average Views per Servings')  # 제목 설정
plt.xlabel('Servings')
plt.ylabel('Average Views')
plt.xticks(rotation=45)
plt.grid(axis='y')

# 파일 저장
plt.savefig('../img/10000_Average Views per Servings.png')  

# 그래프 보여주기
plt.show()
```

## 조리 단계별 레시피 분포

```python
# 각 리스트의 길이를 계산하고 그 중 최대 길이를 구함
max_length = df['cooking_order'].apply(len).max()

print("최대 길이:", max_length)

# 최대 길이와 일치하는 행을 필터링
max_length_row = df[df['cooking_order'].apply(len) == max_length]

max_length_row
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 각 리스트의 길이를 계산
df['cooking_order_length'] = df['cooking_order'].apply(len)

# 길이에 따른 빈도수 계산
length_counts = df['cooking_order_length'].value_counts().sort_index()

# 빈도수 그래프 그리기
plt.figure(figsize=(20, 6))
length_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Recipes by Number of Cooking Steps', fontsize=16)
plt.xlabel('Number of Cooking Steps', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y')

# 파일 저장
plt.savefig('../img/10000_Distribution of Recipes by Number of Cooking Steps.png')  

plt.show()
```

## 조리 단계 수에 따른 레시피의 박스 플롯

```python
# 각 리스트의 길이를 계산
df['cooking_order_length'] = df['cooking_order'].apply(len)

# 빈도수 데이터 프레임 생성
length_counts = df['cooking_order_length']

# 박스 플롯 그리기
plt.figure(figsize=(10, 6))
plt.boxplot(length_counts, vert=False)
plt.title('Box Plot of Recipes by Number of Cooking Steps', fontsize=16)
plt.xlabel('Number of Cooking Steps', fontsize=12)
plt.grid(axis='x')

# 파일 저장
plt.savefig('../img/10000_Box_Plot_of_Recipes_by_Number_of_Cooking_Steps.png')

plt.show()
```

## 조리 단계의 갯수별 평균 조회수

```python
# 1. 조리 단계의 개수 계산
df['step_count'] = df['cooking_order'].apply(len)

# 2. 조리 단계 개수별 조회 수의 평균 계산
average_views_by_steps = df.groupby('step_count')['views'].mean()

# 3. 그래프 그리기
plt.figure(figsize=(20, 6))
average_views_by_steps.plot(kind='bar', color='lightcoral')
plt.title('Average Views by Number of Cooking Steps')
plt.xlabel('Number of Cooking Steps')
plt.ylabel('Average Views')
plt.xticks(rotation=0)  # x축 눈금 회전 각도 설정

# 파일 저장
plt.savefig('../img/10000_Average Views by Number of Cooking Steps.png')  

plt.show()
```

## 재료 갯수별 레시피 분포

```python
# 각 레시피의 재료 개수를 계산
df['ingredient_count'] = df['ingredients'].apply(len)

# 재료 개수에 따른 빈도수 계산
ingredient_counts = df['ingredient_count'].value_counts().sort_index()

# 빈도수 그래프 그리기
plt.figure(figsize=(20, 6))
ingredient_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Recipes by Number of Ingredients', fontsize=16)
plt.xlabel('Number of Ingredients', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y')

# 파일 저장
plt.savefig('../img/10000_Distribution_of_Recipes_by_Number_of_Ingredients.png')

plt.show()
```

## 재료 갯수별 레시피 분포 박스플롯

```python
# 박스플롯 그리기
plt.figure(figsize=(10, 6))
plt.boxplot(df['ingredient_count'], vert=False)
plt.title('Boxplot of Recipe Distribution by Number of Ingredients', fontsize=16)
plt.xlabel('Number of Ingredients', fontsize=12)
plt.grid(axis='x')

# 파일 저장
plt.savefig('../img/10000_Box_Plot_of_Recipe_Distribution_by_Number_of_Ingredients.png')

plt.show()
```

## 재료 갯수별 평균 조회수

```python
# 각 리스트의 길이를 계산하고 그 중 최대 길이를 구함
max_length = df['ingredients'].apply(len).max()

print("최대 길이:", max_length)

# 최대 길이와 일치하는 행을 필터링
max_length_row = df[df['ingredients'].apply(len) == max_length]

max_length_row
```

```python
# 1. 재료의 개수 계산
df['ingredient_count'] = df['ingredients'].apply(len)

# 2. 재료 개수별 조회 수의 평균 계산
average_views = df.groupby('ingredient_count')['views'].mean()

# 3. 그래프 그리기
plt.figure(figsize=(20, 6))
average_views.plot(kind='bar', color='lightcoral')
plt.title('Average Views by Number of Ingredients')
plt.xlabel('Number of Ingredients')
plt.ylabel('Average Views')
plt.xticks(rotation=0)  # x축 눈금 회전 각도 설정

# 파일 저장
plt.savefig('../img/10000_Average Views by Number of Ingredients.png')  

plt.show()
```
